```python
import pandas as pd

# on charge le fichier avec le train
df = pd.read_csv("../Data/processed/train.csv")

# on enleve les lignes qui nous semble inutile
df = df[df['niveau_pollution'] != 'pas de données']
df = df[['Nom de la Station', 'Nom de la ligne']].drop_duplicates()

# on Groupe par ligne pour reconstituer les connections
lines = df.groupby('Nom de la ligne')['Nom de la Station'].apply(list)

# on genere  les connexions directes sur chaque ligne
connections = []
for stations in lines:
    for i in range(len(stations) - 1):
        s1 = stations[i]
        s2 = stations[i + 1]
        connections.append((s1, s2))
        connections.append((s2, s1))  #on detecte les liens B->A pour les supprimer car ce son des doublons

# On constuit la bdd
df_conn = pd.DataFrame(connections, columns=['station1', 'station2'])
df_conn['trajet_existe'] = True

# on utilise la ligne pour supprimer les doublons
df_conn_reduit = df_conn[df_conn['station1'] < df_conn['station2']]

#csv
df_conn_reduit.to_csv("../Data/processed/graphe_stations_reduit.csv", index=False)

print("Fichier généré : graphe_stations_reduit.csv")
print("Nombre de connexions uniques :", len(df_conn_reduit))
```

```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic

# Chargement des fichiers
pollution_df = pd.read_csv("../Data/processed/stations_qgis.csv")
connexion_df = pd.read_csv("../Data/processed/graphe_stations_reduit.csv")

# Supprimer les doublons
unique_pollution_df = pollution_df.drop_duplicates(subset=['Nom de la Station', 'Nom de la ligne'])

# Dictionnaire station → infos pollution / coordonnées
station_info = unique_pollution_df.groupby('Nom de la Station').first()[['stop_lat', 'stop_lon', 'niveau_pollution']].to_dict('index')

# Dictionnaire station → lignes de métro
station_lignes = unique_pollution_df.groupby('Nom de la Station')['Nom de la ligne'].apply(set).to_dict()

# Fonction pour calculer le temps de trajet
def compute_travel_time(row):
    if row['trajet_existe'] != True:
        return None
    s1, s2 = row['station1'], row['station2']
    if s1 in station_info and s2 in station_info:
        coords1 = (station_info[s1]['stop_lat'], station_info[s1]['stop_lon'])
        coords2 = (station_info[s2]['stop_lat'], station_info[s2]['stop_lon'])
        distance_km = geodesic(coords1, coords2).km
        time_hours = distance_km / 65
        return round(time_hours * 60, 2)
    return None

# Fonction pour trouver la ligne de métro utilisée
def get_common_line(row):
    s1, s2 = row['station1'], row['station2']
    lignes1 = station_lignes.get(s1, set())
    lignes2 = station_lignes.get(s2, set())
    common = lignes1.intersection(lignes2)
    if common:
        return list(common)[0]  # ou ", ".join(common) si plusieurs
    return "inconnue"

# Ajout des colonnes
connexion_df['temps_trajet_min'] = connexion_df.apply(compute_travel_time, axis=1)
connexion_df['pollution_station1'] = connexion_df['station1'].map(lambda s: station_info.get(s, {}).get('niveau_pollution', 'inconnue'))
connexion_df['pollution_station2'] = connexion_df['station2'].map(lambda s: station_info.get(s, {}).get('niveau_pollution', 'inconnue'))

connexion_df['lat_station1'] = connexion_df['station1'].map(lambda s: station_info.get(s, {}).get('stop_lat', np.nan))
connexion_df['lon_station1'] = connexion_df['station1'].map(lambda s: station_info.get(s, {}).get('stop_lon', np.nan))
connexion_df['lat_station2'] = connexion_df['station2'].map(lambda s: station_info.get(s, {}).get('stop_lat', np.nan))
connexion_df['lon_station2'] = connexion_df['station2'].map(lambda s: station_info.get(s, {}).get('stop_lon', np.nan))

# Ligne de métro commune
connexion_df['ligne_metro'] = connexion_df.apply(get_common_line, axis=1)

# Export du fichier enrichi
output_path = "../Data/processed/trajets_stations_pollution.csv"
connexion_df.to_csv(output_path, index=False)
print(f"Fichier enrichi exporté avec coordonnées + ligne de métro : {output_path}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Chargement du fichier CSV contenant les trajets, temps, et pollution
df = pd.read_csv("../Data/processed/trajets_stations_pollution.csv")

# Mapping de la pollution textuelle vers un niveau numérique
pollution_niveau = {
    'pollution faible': 1,
    'pollution moyenne': 2,
    'pollution élevée': 3
}

# Création du graphe dirigé pondéré
G = nx.Graph()

# Ajout des arêtes avec le temps de trajet et pollution moyenne comme attributs
for _, row in df.iterrows():
    if row['trajet_existe'] == True and pd.notnull(row['temps_trajet_min']):
        try:
            pollution1 = float(row['pollution_station1'])
            pollution2 = float(row['pollution_station2'])
            pollution_moyenne = np.mean([pollution1, pollution2])
        except:
            pollution_moyenne = np.nan  # en cas de valeur "inconnue"
        
        G.add_edge(row['station1'], row['station2'], 
                   temps=row['temps_trajet_min'], 
                   pollution=pollution_moyenne)
```

```python
def chemin_min_pollution(G, depart, arrivee, temps_max):
    from heapq import heappush, heappop

    # Vérifications de base
    if depart not in G.nodes:
        return f"Station de départ '{depart}' introuvable dans le graphe."
    if arrivee not in G.nodes:
        return f"Station d'arrivée '{arrivee}' introuvable dans le graphe."

    try:
        _ = nx.shortest_path(G, source=depart, target=arrivee)
    except nx.NetworkXNoPath:
        return f"Aucun chemin possible entre '{depart}' et '{arrivee}'."

    # File de priorité : (pollution cumulée, temps cumulé, station actuelle, chemin parcouru)
    file = [(0, 0, depart, [])]

    # Visites mémorisées sous forme (station, temps minimal atteint)
    visited = {}

    while file:
        pollution_cumulee, temps_cumule, current, chemin = heappop(file)
        chemin = chemin + [current]

        # Si déjà visité avec un temps plus court, on ignore
        if current in visited and visited[current] <= temps_cumule:
            continue
        visited[current] = temps_cumule

        # Si on atteint la destination dans les temps
        if current == arrivee and temps_cumule <= temps_max:
            pollution_moyenne = pollution_cumulee / len(chemin)
            return {
                'chemin': chemin,
                'temps_total': round(temps_cumule, 2),
                'pollution_moyenne': round(pollution_moyenne, 2)
            }

        # Parcours des voisins
        for voisin in G.neighbors(current):
            data = G[current][voisin]

            # Gestion de la pollution manquante
            pollution = data['pollution'] if not pd.isna(data['pollution']) else 999

            nouveau_temps = temps_cumule + data['temps']
            if nouveau_temps > temps_max:
                continue

            heappush(file, (
                pollution_cumulee + pollution,
                nouveau_temps,
                voisin,
                chemin
            ))

    return f"Aucun chemin ne respecte la contrainte de temps maximale de {temps_max} minutes entre '{depart}' et '{arrivee}'."
```

```python
# Exemple avec affichage du résultat
resultat = chemin_min_pollution(G, "République", "Gare de Lyon", 20)
if isinstance(resultat, dict):
    display(pd.DataFrame([resultat]))
else:
    print(resultat)
```

```python
import networkx as nx
import pandas as pd

# Rechargement du fichier CSV
df_proposition = pd.read_csv("../Data/processed/trajets_stations_pollution.csv")

# Mapping pollution texte → niveau
pollution_niveau = {
    'pollution faible': 1,
    'pollution moyenne': 2,
    'pollution élevée': 3
}

# Création du graphe pondéré avec pollution intégrée
G_opt = nx.Graph()

for _, row in df_proposition.iterrows():
    if row['trajet_existe']:
        s1, s2 = row['station1'], row['station2']
        p1 = pollution_niveau.get(str(row['pollution_station1']).strip().lower(), 999)
        p2 = pollution_niveau.get(str(row['pollution_station2']).strip().lower(), 999)
        temps = row['temps_trajet_min']
        G_opt.add_node(s1, pollution=p1)
        G_opt.add_node(s2, pollution=p2)
        G_opt.add_edge(s1, s2, temps=temps)

# Fonction principale de recherche optimisée sous contrainte de pollution
def tester_chemins_pollution_optimise(station_depart, station_arrivee):
    """
    Teste les chemins optimaux en temps entre deux stations sous différents seuils de pollution.
    Retourne un tableau avec :
    - Seuil utilisé
    - Chemin trouvé ou non
    - Chemin optimal (minimisant le temps)
    - Pollution sur le trajet
    - Temps total estimé
    """
    resultats = []

    for seuil in [1, 2, 3]:
        valid_nodes = [n for n, d in G_opt.nodes(data=True) if d['pollution'] <= seuil]
        H = G_opt.subgraph(valid_nodes).copy()

        try:
            chemin = nx.shortest_path(H, source=station_depart, target=station_arrivee, weight="temps")
            pollution_chemin = [G_opt.nodes[n]['pollution'] for n in chemin]
            temps_total = sum(G_opt.edges[chemin[i], chemin[i+1]]['temps'] for i in range(len(chemin)-1))
            pollution_textuelle = ["faible" if p == 1 else "moyenne" if p == 2 else "élevée" for p in pollution_chemin]
            description_seuil = (
                "faible" if seuil == 1 else
                "faible ou moyenne" if seuil == 2 else
                "faible, moyenne ou élevée"
            )
            resultats.append({
                "Seuil de pollution autorisé": description_seuil,
                "Chemin trouvé": "Oui",
                "Trajet optimal (temps)": " -> ".join(chemin),
                "Pollution sur le trajet": pollution_textuelle,
                "Temps total (min)": round(temps_total, 2)
            })
        except (nx.NetworkXNoPath, nx.NodeNotFound):
            description_seuil = (
                "faible" if seuil == 1 else
                "faible ou moyenne" if seuil == 2 else
                "faible, moyenne ou élevée"
            )
            resultats.append({
                "Seuil de pollution autorisé": description_seuil,
                "Chemin trouvé": "Non",
                "Trajet optimal (temps)": "Aucun",
                "Pollution sur le trajet": "N/A",
                "Temps total (min)": "N/A"
            })

    pd.set_option('display.max_colwidth', None)
    display(pd.DataFrame(resultats))
```

```python
# Exemple d'utilisation
resultats = tester_chemins_pollution_optimise("Gare de Lyon", "Bastille")
if isinstance(resultats, list) and all(isinstance(r, dict) for r in resultats):
    display(pd.DataFrame(resultats))
else:
    print(resultats)
```

```python
import pandas as pd
import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display

# Préparation des stations disponibles
stations_list = sorted(set(G_txt.nodes))

# Widgets
station_depart_widget = widgets.Dropdown(options=stations_list, description="Départ")
station_arrivee_widget = widgets.Dropdown(options=stations_list, description="Arrivée")
fonction_widget = widgets.RadioButtons(
    options=["tester_chemins_pollution", "chemin_min_pollution"],
    description="Fonction:"
)
temps_max_widget = widgets.IntSlider(value=15, min=1, max=60, step=1, description="Temps max")

def afficher_resultats(depart, arrivee, fonction, temps_max):
    import itertools
    from matplotlib import cm

    print(f"🟢 Départ : {depart}   🔵 Arrivée : {arrivee}")
    print(f"📌 Fonction choisie : {fonction}")

    fig = go.Figure()
    chemins_affiches = False

    if fonction == "tester_chemins_pollution":
        resultats = tester_chemins_pollution(depart, arrivee)
        display(pd.DataFrame(resultats))

        # Palette de couleurs
        cmap = cm.get_cmap('tab10')
        couleurs = itertools.cycle([
            f'rgb({int(r*255)},{int(g*255)},{int(b*255)})'
            for r, g, b, _ in cmap(np.linspace(0, 1, 10))
        ])

        for resultat in resultats:
            if resultat["Chemin trouvé"] == "Oui":
                chemin = resultat["Trajet"].split(" -> ")
                couleur = next(couleurs)

                # Tracé du chemin
                for i in range(len(chemin) - 1):
                    n1, n2 = chemin[i], chemin[i+1]
                    lat1, lon1 = G_txt.nodes[n1].get('lat', None), G_txt.nodes[n1].get('lon', None)
                    lat2, lon2 = G_txt.nodes[n2].get('lat', None), G_txt.nodes[n2].get('lon', None)
                    if lat1 and lon1 and lat2 and lon2:
                        fig.add_trace(go.Scattermapbox(
                            lat=[lat1, lat2],
                            lon=[lon1, lon2],
                            mode='lines',
                            line=dict(width=2, color=couleur),
                            showlegend=False
                        ))

                # Points
                coords = [(G_txt.nodes[n].get('lat', None), G_txt.nodes[n].get('lon', None)) for n in chemin]
                fig.add_trace(go.Scattermapbox(
                    lat=[lat for lat, lon in coords if lat],
                    lon=[lon for lat, lon in coords if lon],
                    mode='markers+text',
                    text=chemin,
                    marker=dict(size=9, color=couleur),
                    name='Chemin'
                ))

                chemins_affiches = True

    else:
        res = chemin_min_pollution(G_txt, depart, arrivee, temps_max)
        if isinstance(res, str):
            print(res)
            return
        display(pd.DataFrame([res]))
        chemin = res['chemin']

        # Tracé du chemin
        for i in range(len(chemin) - 1):
            n1, n2 = chemin[i], chemin[i+1]
            lat1, lon1 = G_txt.nodes[n1].get('lat', None), G_txt.nodes[n1].get('lon', None)
            lat2, lon2 = G_txt.nodes[n2].get('lat', None), G_txt.nodes[n2].get('lon', None)
            if lat1 and lon1 and lat2 and lon2:
                fig.add_trace(go.Scattermapbox(
                    lat=[lat1, lat2],
                    lon=[lon1, lon2],
                    mode='lines',
                    line=dict(width=3, color='blue'),
                    showlegend=False
                ))

        coords = [(G_txt.nodes[n].get('lat', None), G_txt.nodes[n].get('lon', None)) for n in chemin]
        fig.add_trace(go.Scattermapbox(
            lat=[lat for lat, lon in coords if lat],
            lon=[lon for lat, lon in coords if lon],
            mode='markers+text',
            text=chemin,
            marker=dict(size=10, color='red'),
            name='Chemin optimal'
        ))
        chemins_affiches = True

    if chemins_affiches:
        fig.update_layout(
            mapbox_style="carto-positron",
            mapbox_zoom=11,
            mapbox_center={"lat": coords[0][0], "lon": coords[0][1]} if coords[0][0] else {"lat": 48.85, "lon": 2.35},
            height=600,
            margin={"r": 0, "t": 30, "l": 0, "b": 0},
            title="Trajets sur le graphe"
        )
        fig.show()
    else:
        print("❗ Aucun chemin visualisable trouvé pour affichage cartographique.")


# Interface interactive
ui = widgets.VBox([
    station_depart_widget,
    station_arrivee_widget,
    fonction_widget,
    temps_max_widget
])

out = widgets.interactive_output(
    afficher_resultats,
    {
        'depart': station_depart_widget,
        'arrivee': station_arrivee_widget,
        'fonction': fonction_widget,
        'temps_max': temps_max_widget
    }
)

display(ui, out)
```
